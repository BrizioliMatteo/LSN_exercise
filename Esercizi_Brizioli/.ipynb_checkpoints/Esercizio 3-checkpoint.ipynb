## <span style="color:blue">  Numerical exercises 3</span>

La teoria di Black-Scholes assume che l'evoluzione temporale del "asset price" sia stocastica e che segua un geometric Brownian motion(GBM) con rate di interesse $r$ a $\textit{risk-free}$  e volatilità $\sigma$.



Possiamo pertanto scrivere l'equazione differenziale stocastica per l'$\textit{asset price}$:
\begin{cases}
dS(t)=\mu S(t)dt+\sigma S(t)dW(t)\\
t\geq 0,S(0)
\end{cases}
Nel caso in cui $\sigma=0$ l'evoluzione è $\textit{risk-less}$, mentre $\mu$ misura il rate di crescita medio dell'$\textit{asset price}$.In pratica $\sigma$ introduce un $\textit{risk}$ nell'evoluzione temporale.

La risoluzione di tale equazione stocastica per
$S\sim GBM(\mu,\sigma^2)$,  $S(0)$, sarà quindi:
    $$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
dove $W(t)\sim\mathcal{N}(0,t)$.

Consideriamo le $\textit{European options}$, che sono contratti tra due parti in cui il writer dell'opzione garantisce all' holder il diritto di comprare ($\textit{call option}$) o vendere ($\textit{put option}$) ad un prezzo prefissato $K$ ($\textit{strike price}$) l'attività finanziara (avente un prezzo attuale $S(t)\textit{ spot price}$).
Inoltre la facoltà di vendere o comprare è limitata alla sola data di scadenza ($\textit{expiring date}$). In altre parole, nel caso di una call option, l'holder si garantisce il diritto di comprare una certa attività finanziaria ad un certo prezzo fissato $K$ al tempo $T$. Valutando il prezzo dell'opzione $S(T)$ al tempo $T$ e quindi il profitto $\max(0,S(T)-K)$ che ne potrebbe ricavare definisce il prezzo iniziale $C(0)$ della $\textit{Call-option}=\langle e^{-rT}\max[0,S(T)-K]\rangle$.

Gli andamenti della call-option e della put-option (valutati campionando la $S(t)$ in maniera diretta o discretizzata) tendono, entro l'incertezza statistica, all'aumentare del numero di blocchi al valore atteso stimato di Black-Scholes.

Come abbiamo già detto l'asset price $S(t)$ segue il $\textit{Geometric Brownian Motion}$ ovvero $\log(S(t))$ è un moto browniano con valore iniziale $\log(S(0))$. 
Inoltre dato he gli incrementi di $W$ sono indipendenti e normalmente distribuiti , la semplice procedura di ricorsione per campionare valori discretizzati di $S\sim GBM(\mu,\sigma^2)$ a $t_1 < t_2 < … < t_n$ è:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    con $Z_i \sim \mathcal{N}(0,1)$ ovvero Gaussiane standard indipendenti.
</p>
Qusto metodo di discretizzazione è esatto e la serie $(S(t_1),\dots,S(t_n))$ che esso produce produce ha una probabilità congiunta $S\sim GBM(\mu,\sigma ^2)$ al tempo $t_1,\dots,t_n$.
</p>
Anche le simulazioni pertanto mostrano che il risultato è lo stesso per un campionamento diretto o discretizzato.
