Il Simulated Annealing è basato su una analogia con l'annealing termico dei materiali, evitando che il sistema rimanga intrappolato in un minimo locale grazie ai criteri di accettazione del Metropolis. L'idea pertanto è quella di interpretare la loss function $L(x)$ come l'energia di un sistema fisico di meccanica statistica e considerare la distribuzione di Boltzmann: $$p(x)=\frac{1}{Z}e^{-\beta L(x)}.$$
A questo punto si lascia variare $\beta$ con il tempo, ossia si considera un un set $\{\beta_1,n_1\}\dots\{\beta_N,n_N\}$ con $\beta_i \in(0,\infty)$ e $n_i\geq0$. Quindi partendo da una configurazione iniziale $x_0$, si eseguono $n_1$ Monte Carlo steps alla temperatura$1/\beta_1$, la configurazione finale per $\{\beta_1,n_1\}$ viene allora utilizzata come configurazione iniziale per la configurazione $\{\beta_2,n_2\}$ e così via  fino a $\{\beta_N,n_N\}$. Essendo ogni processo Monte Carlo ergodico, ci si aspetta che al diminuire di $\beta$ la simulazione finisca in un minimo globale dell'energia.


Negli esercizi il simulated annealing è stato effettuato utilizando il **Geometric Cooling** in cui $\beta_{k+1}:=c\beta_k$ cone $ c=1.1$

By adapting your Genetic Algorithm code, developed during the Numerical Exercise 9, write a C++ code to solve the TSP with a **Simulated Annealing** (SA) algorithm. Apply your code to the optimization of a path among

Show your results via:

Riprendendo il codice dell'algoritmo genetico, si è eliminata la procedura di selezione e il Crossover. Le mosse proposte del processo Monte Carlo sono solamente le Mutazioni con una certa probabilità. Quindi partendo da un individuo si genera proponendo delle mutazioni un nuovo individuo e si valuta la probabilità di accettazione che sarà:
\begin{cases}
e^{-\beta(L(x')-L(x))} & if L(x')>L(x) \\
1 & altrimenti
\end{cases}
e dopo una certa serie di iterazioni a temperatura fissata, si esegue di nuovo un processo Monte Carlo a temperatura diversa partendo però dall'individuo finale della configurazione precedente.

Si sono utilizzati $M=10^5$ step Monte Carlo e $N_t=100$ temperature.

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
import math
import matplotlib.gridspec as gridspec



N_gen_c, I_best_c, = np.loadtxt("Es10.1/Best_circ.dat", usecols=(0,1), delimiter='\t', unpack='true')
N_gen_r, I_best_r, = np.loadtxt("Es10.1/Best_rec.dat", usecols=(0,1), delimiter='\t', unpack='true')


plt.figure(figsize=(20,10))
plt.plot(N_gen_c,I_best_c,label="Circle")
plt.plot(N_gen_r,I_best_r,label="Square")
plt.title(r'$ best path$',fontsize='x-large')
plt.xlabel('#Iterazioni',fontsize='x-large')
plt.ylabel(r'$L^{(1)}_{best}$',fontsize='x-large')
plt.legend(fontsize='x-large')
plt.grid(True)

plt.show()
```

Si nota come dopo poche iterazioni $\sim 40$ il sistema tende al minimo.

**MPI- Message Passing Interface** si basa sul modello message-passsing. In particolare esso pone un insieme di processi che hanno soltanto della memoria locale ma che sono capaci di parlarsi, di comunicare, mandando e ricevendo messaggio. Una carretteristica importante del message-passing model è che i dati transferiti dalla memoria locale di un processo alla memoria locale di un altrorichiede operazioni che siano eseguite su entrambe i processi. In altre parole, dato che in tale modello i processi eseguiti in parallelo hanno differente indirizzi di memoria, le comunicazioni avvengono quando una parte dello spazio di memoria di un processo è copiato nello spazio di memoria di un altro processo. Questa operazione è cooperativa e si ha soltanto quando il primo processo esegue un *send* e il secondo esegue un'operazione di *receive*.


Nel codice implementato si sono utilizzate solamente comunicazioni di tipo collettivo. In particolare, oltre ad avere inizializzato il processo, si è utilizzata la funzione *MPI_Bcast* per settare tutti i parametri di input,   per tutti i processi leggendo così una sola volta i file, tuttavia cambiando i primes per ogni generatore random. Infine si è utilizzato *MPI_Allreduce* per trovare la funziione costo minima tra tutti processi conservandone il valore e il numero del processo.

Per testare la funzionalità del parallelo si è eseguita la simulazione con un numero di città $N_c=100$, con un numero di step Monte Carlo per ogni temperatura di $N_{step}=5000.$

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
import math
import matplotlib.gridspec as gridspec



N_it0, I_best0 = np.loadtxt("Es10.02/Data/Best_0_1.dat", usecols=(0,1), delimiter='\t', unpack='true')
N_it1, I_best1 = np.loadtxt("Es10.02/Data/Best_1_1.dat", usecols=(0,1), delimiter='\t', unpack='true')
N_it2, I_best2 = np.loadtxt("Es10.02/Data/Best_2_1.dat", usecols=(0,1), delimiter='\t', unpack='true')
N_it3, I_best3 = np.loadtxt("Es10.02/Data/Best_3_1.dat", usecols=(0,1), delimiter='\t', unpack='true')
N_it4, I_best4 = np.loadtxt("Es10.02/Data/Best_4_1.dat", usecols=(0,1), delimiter='\t', unpack='true')
N_it5, I_best5 = np.loadtxt("Es10.02/Data/Best_5_1.dat", usecols=(0,1), delimiter='\t', unpack='true')



N_its0, I_best0s = np.loadtxt("Es10.02/Data/Best_0_0.dat", usecols=(0,1), delimiter='\t', unpack='true')
N_its1, I_best1s = np.loadtxt("Es10.02/Data/Best_1_0.dat", usecols=(0,1), delimiter='\t', unpack='true')
N_its2, I_best2s = np.loadtxt("Es10.02/Data/Best_2_0.dat", usecols=(0,1), delimiter='\t', unpack='true')
N_its3, I_best3s = np.loadtxt("Es10.02/Data/Best_3_0.dat", usecols=(0,1), delimiter='\t', unpack='true')
N_its4, I_best4s = np.loadtxt("Es10.02/Data/Best_4_0.dat", usecols=(0,1), delimiter='\t', unpack='true')
N_its5, I_best5s = np.loadtxt("Es10.02/Data/Best_5_0.dat", usecols=(0,1), delimiter='\t', unpack='true')


G = gridspec.GridSpec(2,1)
f = plt.figure(figsize=(20,30))
f.subplots_adjust(hspace=0.4, wspace=0.4)
fig_1=plt.subplot(G[0, 0])
fig_1.plot(N_it0,I_best0,label="Rank 0")
fig_1.plot(N_it1,I_best1,label="Rank 1")
fig_1.plot(N_it2,I_best2,label="Rank 2")
fig_1.plot(N_it3,I_best3,label="Rank 3")
fig_1.plot(N_it4,I_best4,label="Rank 4")
fig_1.plot(N_it5,I_best5,label="Rank 5")
fig_1.set_title(r'100 cities randomly placed on a circumference',fontsize='x-large')
fig_1.set_xlabel('#Iterazione',fontsize='x-large')
fig_1.set_ylabel('$L^{(1)}_{best}$',fontsize='x-large')
fig_1.legend(fontsize='x-large')
fig_1.grid(True)

fig_2=plt.subplot(G[1, 0])
fig_2.plot(N_its0,I_best0s,label="Rank 0")
fig_2.plot(N_its1,I_best1s,label="Rank 1")
fig_2.plot(N_its2,I_best2s,label="Rank 2")
fig_2.plot(N_its3,I_best3s,label="Rank 3")
fig_2.plot(N_its4,I_best4s,label="Rank 4")
fig_2.plot(N_its5,I_best5s,label="Rank 5")
fig_2.set_title(r'100 cities randomly placed inside a square',fontsize='x-large')
fig_2.set_xlabel('#Iterazione',fontsize='x-large')
fig_2.set_ylabel(r'$L^{(1)}_{best}$',fontsize='x-large')
fig_2.legend(fontsize='x-large')
fig_2.grid(True)

plt.show()
```
