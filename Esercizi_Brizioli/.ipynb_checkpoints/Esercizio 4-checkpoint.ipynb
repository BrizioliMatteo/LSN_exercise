## <span style="color:blue">  Numerical exercises 4</span>

Molecular dynamics simulation è una tecinca per stimare le proprietà di equilibrio e trasporto di sistemi calssici a molti-corpi. Dove per classico si intende che il moto delle particelle obbedisce alle leggi della meccanica classica.<p>
In una simulazione di dimanica molecolare si segue un approccio simile a quello sperimentale: per prima cosa si prepara il campione considerando un sistema modello di N particelle e si risolvono le equazioni del moto per tale sistema fino a quando le proprietà del sistema non variano più nel tempo. Dopo l'equilibrazione, si effettuano le misure.<p>

Quindi l'evoluzione temporale del sistema di particelle interagenti viene effettuato integrando le loro equazioni del moto con specifiche condizioni al contorno. In particolare le simulazioni vengono effettute a $(N,V,E)$ costante ovvero si considera un ensemble microcanonico.

Pertanto i punti fondamentali di una simulazione di Molecular Dynamics sono:<p>
$\textbf{Inizializzazione}$:Per iniziare la simulazione vengono assegnate le posizizioni e le velocità a tutte le particelle del sistema a seconda della struttura che si sta cercando di simulare. Occorre tuttavia che le particelle non si trovino in posizioni che abbiano un certo overlap fra di loro.<p>
$\textbf{Calcolo della Forza}$: a questo punto occorre calcolare la forza agente su ogni singola particella. Nel nostro modello consideriamo un'interazione di coppia di tipo Lennard-Jones. Se si considerano soltanto interazioni tra una particella e la più vicina immagine di un'altra particella (minimum image convention), occorre calcolare $N(N-1)/2$ termini, un'altra aprossimazione è quella di imporre un cut-off sferico, limitando a calcolare il contributo al potenziale e alle forze sia dovuto alle particelle più vicine. Con questo cut-off tuttavia le proprietà termodinamiche del sistema non sono più esatte. Come condizione su $r_c$ è che deve essere minore di $L/2$ dove $L$ è il lato della scatola dove è contenuto il sistema.<p>
$\textbf{Integrazione delle equazioni del moto}$: Una volta stimate le forze tra le particelle, si possono integrare le equazioni di Newton. Per farlo si è utilizzato l'algoritmo di Verlet che non utilizza la velocità per stimare le nuove posizioni:$$r(t+\delta t)=2r(t)-r(t-\delta t)+\frac{f(t)}{m}\delta t^2+O(\delta t^4)$$ da qui si possono ricavare anche le velocità:$$v(t)=\frac{r(t+\delta t)-r(t-\delta t)}{2\delta t}+O(\delta t^2).$$ Viene utilizzato l'algoritmo di Verlet perchè sebbene abbia una conservazione del'energia per tempi moderatamente brevi presenta invece un piccola deriva dell'energia a lungo termine. 

### Exercise 04

By means of your upgraded MD code, equilibrate and <span style="color:red">perform MD simulations via a Lennard-Jones model</span> of Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) and Krypton ($\sigma = 0.364$ nm, $\epsilon/k_B = 164$ K, $m=83.798$ amu) in the following conditions:
1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

and show in pictures the obtained average values and uncertainties for the total (internal) energy per particle, $E/N$, the potential energy per particle, $U/N$, the kinetic energy per particle, $K/N$, the pressure $P$ and the temperature, $T$

Per prima cosa si è effettuata l'equilibrazione del sistema per le tre diverse fasi del sistema solida, liquida e gassosa.<p>
L'algortimo come abbiamo detto lavora ad energia costante, tuttavia vogliamo studiare il sistema di particelle all'equilibrio per una certa temperatura $T$ fissata. Pertanto occorrerà riequilibrare il sistema riemponendo al sistema la temperatura iniziale.<n>
Per far questo, abbiamo migliorato quindi il codice dando la possibilità di rininiziare da una certa configurazione.

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
import math

Epot_s = np.loadtxt("Es4/Molecular_dynamics/Instant/output_epot_solid.dat", usecols=(0), unpack='true')
Ekin_s = np.loadtxt("Es4/Molecular_dynamics/Instant/output_ekin_solid.dat", usecols=(0), unpack='true')
Etot_s = np.loadtxt("Es4/Molecular_dynamics/Instant/output_etot_solid.dat", usecols=(0), unpack='true')
Temp_s = np.loadtxt("Es4/Molecular_dynamics/Instant/output_temp_solid.dat", usecols=(0), unpack='true')
x_s= np.arange(len(Epot_s))

Epot_l= np.loadtxt("Es4/Molecular_dynamics/Instant/output_epot_liquid.dat", usecols=(0), unpack='true')
Ekin_l= np.loadtxt("Es4/Molecular_dynamics/Instant/output_ekin_liquid.dat", usecols=(0), unpack='true')
Etot_l= np.loadtxt("Es4/Molecular_dynamics/Instant/output_etot_liquid.dat", usecols=(0), unpack='true')
Temp_l= np.loadtxt("Es4/Molecular_dynamics/Instant/output_temp_liquid.dat", usecols=(0), unpack='true')
x_l= np.arange(len(Epot_l))

Epot_g= np.loadtxt("Es4/Molecular_dynamics/Instant/output_epot_gas.dat", usecols=(0), unpack='true')
Ekin_g= np.loadtxt("Es4/Molecular_dynamics/Instant/output_ekin_gas.dat", usecols=(0), unpack='true')
Etot_g= np.loadtxt("Es4/Molecular_dynamics/Instant/output_etot_gas.dat", usecols=(0), unpack='true')
Temp_g= np.loadtxt("Es4/Molecular_dynamics/Instant/output_temp_gas.dat", usecols=(0), unpack='true')
x_g= np.arange(len(Epot_g))

f, axarr = plt.subplots(3,1,figsize=(10,15))
f.subplots_adjust(hspace=0.4, wspace=0.4)

axarr[0].plot(x_s,Epot_s,'r',label="E_pot")
axarr[0].plot(x_s,Ekin_s,'g',label="E_kin")
axarr[0].plot(x_s,Etot_s,'b',label="E_tot")
axarr[0].plot(x_s,Temp_s,'y',label="Temp")
axarr[0].set_title('Solid Phase',fontsize='x-large')
axarr[0].set_xlabel('Step(1/10)',fontsize='x-large')
axarr[0].set_ylabel(r'$[A U]$',fontsize='x-large')
axarr[0].legend(fontsize='x-large')
axarr[0].grid(True)

axarr[1].plot(x_l,Epot_l,'r',label="E_pot")
axarr[1].plot(x_l,Ekin_l,'g',label="E_kin")
axarr[1].plot(x_l,Etot_l,'b',label="E_tot")
axarr[1].plot(x_l,Temp_l,'y',label="Temp")
axarr[1].set_title('Liquid Phase',fontsize='x-large')
axarr[1].set_xlabel('Step (1/10)',fontsize='x-large')
axarr[1].set_ylabel(r'$[A U]$',fontsize='x-large')
axarr[1].legend(fontsize='x-large')
axarr[1].grid(True)

axarr[2].plot(x_g,Epot_g,'r',label="E_pot")
axarr[2].plot(x_g,Ekin_g,'g',label="E_kin")
axarr[2].plot(x_g,Etot_g,'b',label="E_tot")
axarr[2].plot(x_g,Temp_g,'y',label="Temp")
axarr[2].set_title('Gas Phase',fontsize='x-large')
axarr[2].set_xlabel('Step(1/10)',fontsize='x-large')
axarr[2].set_ylabel(r'$[A U]$',fontsize='x-large')
axarr[2].legend(fontsize='x-large')
axarr[2].grid(True)
plt.show()
```

Nei grafici sono riportati i valori istantanei dell'energia potenziale,dell'energia cinetica, dell'energia totale e della temperatura tutte valutate in unità ridotte. <n> Ogni $N=10000$ step è stata riempostata la temperatura al sistema, ovvero fatta ripartire la simulazione considerando le configurazioni finali di ognuna e riscalate le velocità per matchare con la nuova temperatura imposta. Come si può notare  la fase solida tra le tre fasi è quella che presenta un minore tempo di equilibrazione. Si nota inoltre che l'energia totale rimane approssimativamente costante e non mostra un drift durante le varie simulazioni di equilibrazione.

Una volta equilibrato il sistema si può procedere alla misurazione delle osservabili d'interesse

```python
sigma=0.34*(10**-9)
e_kb=120.
eps=120.*1.380649*(10**-23)
import matplotlib.gridspec as gridspec


x1s,Epot_ave_s, dev_Epot_s = np.loadtxt("Es4/Molecular_dynamics/Output/ave_epot_solid.out", usecols=(0,1,2), unpack='true')
x2s,Ekin_ave_s, dev_Ekin_s = np.loadtxt("Es4/Molecular_dynamics/Output/ave_ekin_solid.out", usecols=(0,1,2), unpack='true')
x3s,Etot_ave_s, dev_Etot_s = np.loadtxt("Es4/Molecular_dynamics/Output/ave_etot_solid.out", usecols=(0,1,2), unpack='true')
x4s,Temp_ave_s, dev_Temp_s= np.loadtxt("Es4/Molecular_dynamics/Output/ave_temp_solid.out", usecols=(0,1,2), unpack='true')
x5s,Pres_ave_s, dev_Pres_s= np.loadtxt("Es4/Molecular_dynamics/Output/ave_pres_solid.out", usecols=(0,1,2), unpack='true')

G = gridspec.GridSpec(3,2)
f = plt.figure(figsize=(12,15))
f.subplots_adjust(hspace=0.4, wspace=0.4)
f.suptitle('Solid Phase-Argon',fontsize='x-large')

axa0=plt.subplot(G[0, 0])
axa0.errorbar(x1s,Epot_ave_s*eps,yerr=dev_Epot_s*eps,label="E_pot")
axa0.set_title('Energia Potenziale',fontsize='x-large')
axa0.set_xlabel('#Blocks',fontsize='x-large')
axa0.set_ylabel(r'$E_{pot}/N [J]$',fontsize='x-large')
axa0.grid(True)

axa1=plt.subplot(G[0, 1])
axa1.errorbar(x2s,Ekin_ave_s*eps,yerr=dev_Ekin_s*eps,label="E_kin")
axa1.set_title('Energia Cinetica',fontsize='x-large')
axa1.set_xlabel('#Blocks',fontsize='x-large')
axa1.set_ylabel(r'$E_{kin}/N [J]$',fontsize='x-large')
axa1.grid(True)
  
axa2=plt.subplot(G[1,:])
axa2.errorbar(x3s,Etot_ave_s*eps,yerr=dev_Etot_s*eps,label="E_tot")
axa2.set_title('Energia Totale',fontsize='x-large')
axa2.set_xlabel('#Blocks',fontsize='x-large')
axa2.set_ylabel(r'$E_{tot}/N [J]$',fontsize='x-large')
axa2.grid(True)

axa3=plt.subplot(G[2, 0])
axa3.errorbar(x4s,Temp_ave_s*e_kb,yerr=dev_Temp_s*e_kb,label="Temp")
axa3.set_title('Temperatura',fontsize='x-large')
axa3.set_xlabel('#Blocks',fontsize='x-large')
axa3.set_ylabel(r'$T [K]$',fontsize='x-large')
axa3.grid(True)

axa4=plt.subplot(G[2, 1])
axa4.errorbar(x5s,Pres_ave_s*eps/(sigma**3),yerr=dev_Pres_s*eps/(sigma**3),label="Pres")
axa4.set_title('Pressione',fontsize='x-large')
axa4.set_xlabel('#Blocks',fontsize='x-large')
axa4.set_ylabel(r'$P [Pa]$',fontsize='x-large')
axa4.grid(True)
plt.show()
```

Per il calocolo statistico delle incertezze e dei valori medi si è utilizzato il data-blocking. Questo perchè in una simulazione di dinamica molecolare il tempo tra due punti successivi della simulazione è di solito troppo breve per ottenere un campionamento statistico indipendente, cioè tali punti in genere sono fortemente correlati. Il raggruppamento dei dati in dati in blocchi determina l'eliminazione di tale correlazione e quindi l'applicabilità del Teorema del Limite Centrale (variabili indipendenti) e quindi permettendo la stima statistica delle grandezze.
