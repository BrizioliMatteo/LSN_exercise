Il Pseudo-Random Number generator è un algoritmo capace di campionare una sequenza di random variabili indipendenti uniformemente nell'intervallo $[0,1)$. Pertanto, il risultato di tale algoritmo è una sequenza
\begin{equation}
(x_1,\dots,x_n)
\end{equation}
con $0<x_i<1$,di realizzazioni di $n$ random variabili uniformente distribuite nell'intervallo $[0,1)$:
\begin{equation}
(X_1,\dots,X_n).
\end{equation}


Una applicazione molto importante del sampling di distribuzioni di Random variabili e quindi di distribuzioni di probabilità l'integrazione Monte Carlo.
Cosideriamo infatti l'integrale nella sua forma più generica:
\begin{equation}
I=\int_D d\textbf{x} f(\textbf{x})p(\textbf{x})
\end{equation}
Dove $D \subset \Re^{d}$, $f$ è una funzione misurabile e $p$ una certa distribuzione di probabilità su $D$ tale che:
\begin{equation}
p(\textbf{x})\geq 0, \int_D p(\textbf{x})=1.
\end{equation}
Quindi, data una certa random variabile $X$ con una certa distribuzione di probabilità $p$ vale che:
\begin{equation}
  I=E[f(X)] 
 \end{equation}
\begin{equation}
  \int_D d\textbf{x} (f(\textbf{x})-I)^{2}p(\textbf{x})=Var(f(X))
 \end{equation}
La legge dei grandi numeri e il teorema del limite centrale, avendo che $E[f(X)]$ e $Var(f(X))$ sono finiti, garantisce che data una sequenza $\{X_i\}_{i=1}^{\infty}$, si ha che:
\begin{equation}
I=lim_{n\rightarrow\infty}\frac{1}{n}\sum_{i=1}^nf(X_i)
\end{equation}
Perciò se siamo capaci di campionare la distribuzione di probabilità $p$, cioè di generare delle realizzazioni in $D$ di $\{X_i\}_{i=1}^n$ possiamo calcolare l'integrale:
\begin{equation}
I\simeq\frac{1}{n}\sum_{i=1}^nf(X_i)
\end{equation}
e la relativa deviazione standard.

Un metodo per stimare l'incertezza statistica è il data blocking, utile nel caso in cui non ho random variabili che sono statisticamente dipendenti e quindi non è valido il teorema del limite centrale. Consideriamo un'integrazione Monte Carlo che utlizza un numero $M$ di estrazioni e dividiamo tali estrazioni in $N$ blocchi con $N<M$. In ogni blocco, si possono utilizzare $M/N$ estrazioni Monte Carlo per valutare $I$ e quindi abbiamo che:
\begin{equation}
\frac{\sigma}{\sqrt{N-1}}=\sqrt{\frac{1}{\sqrt{N-1}}\bigg[\frac{1}{N}\sum_{j=1}^N I_j^2-\bigg(\frac{1}{N}\sum_{i=1}^{N}I_j\bigg)^2\bigg]}
\end{equation}

Considerando il teorema del limite centrale, in questo caso la deviazione standard fatta con il blocking risulta essere uguale alla deviazione standard su tutte l'estrazioni Montecarlo.

3) Divide $[0,1]$ into $M$ identical sub-intervals and implement the $\chi^2$ test. Obviously, the number of expected events observed in each sub-interval after $n$ *throws*, according to a uniform distribution, is $np = n\times 1/M= n/M$. Fix $M=10^2$ and use for $n$ the first $10^4$ pseudo-random numbers, then the successive $10^4$ pseudo-random numbers, and so on ... 100 times. <font color="red">Plot $\chi^2_j$ for $j=1, ..., 100$</font>. 

Valutiamo pertanto il $\chi^2$:
$$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$
Dove $n/M$ non è altro che il valore di aspettazione ovvero sui singoli sotto-intervalli in cui è stato suddiviso l'intervallo di dispersione, dato che stiamo estraendo utilizzando una distribuzione uniforme. Per tanto ci aspettiamo che il che per grandi valori di $n$ in media si ha che $(n_i - n/M)^2 \simeq n/M$ e quindi $\chi^2 \simeq 100$, cioè uguale al numero di sottointervalli.

```python
plt.figure(figsize=(8,8))

x, chi = np.loadtxt("Es1/chi_squared.out", usecols=(0,1), delimiter='\t', unpack='true')
atteso=np.ones(100)*100
plt.plot(x,chi,'o')
plt.plot(x,atteso,'r',linewidth=2.0)
plt.xlabel('Interval',fontsize='x-large')
plt.ylabel(r'$\chi^{2}$',fontsize='x-large')


print(max(abs(atteso-chi)))

plt.show()
```

Il $\chi^2$ su ogni sottointervallo risulta pertanto distributito attorno al valore atteso di $100$  (avendo imposto $M=100$ e $n=10^4$) con un massimo discostamento di $33.2$. Questo ci dice statisticamente che le occorenze sono estratte in maniera uniforme e che seguono la distribuzione di probabilità attesa. Quindi in ultimo che il Pseudo-Random generator, sebbene sia un algoritmo deterministico genera numeri con la probabilità attesa.

```python
def gaus(x,a,x0,sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))


x1,y1= np.loadtxt("Es1.1/Histo_1.txt", usecols=(0,1), delimiter='\t', unpack='true')
x2,y2= np.loadtxt("Es1.1/Histo_2.txt", usecols=(0,1), delimiter='\t', unpack='true')
x3,y3= np.loadtxt("Es1.1/Histo_10.txt", usecols=(0,1), delimiter='\t', unpack='true')
x4,y4= np.loadtxt("Es1.1/Histo_100.txt", usecols=(0,1), delimiter='\t', unpack='true')

par,pcov = optimize.curve_fit(gaus,x4,y4,p0=[1,0.,0.2])

f, axarr = plt.subplots(2,2,figsize=(20,20))
f.subplots_adjust(hspace=0.2, wspace=0.2)

axarr[0, 0].bar(x1, y1, width=0.02)
axarr[0, 0].set_title(r'$N=1$',fontsize='x-large')
axarr[0, 1].bar(x2, y2, width=0.02)
axarr[0, 1].set_title(r'$N=2$',fontsize='x-large')
axarr[1, 0].bar(x3, y3, width=0.02)
axarr[1, 0].set_title(r'$N=10$',fontsize='x-large')
axarr[1, 1].bar(x4, y4, width=0.02,label="Data")
axarr[1, 1].plot(x4,gaus(x4,par[0],par[1],par[2]),'r',label="fit")
axarr[1, 1].legend(fontsize='x-large')
axarr[1, 1].set_title(r'$N=100$',fontsize='x-large')

f.suptitle('Standard Dice',fontsize='x-large')
print("Fit= ", par)
plt.show()
```

Come si può notare la distribuzione della variabile somma, all'aumentare $N$ del numero di Random variabili indipendenti $(x_1,\dots,x_N)$ e identicamente distribuite con valor medio $\langle x_1 \rangle=\dots=\langle x_n\rangle=\mu$ e varianza $\langle x_1^2 \rangle-\langle x_1\rangle^2=\dots=\langle x_N^2 \rangle-\langle x_N \rangle^2=\sigma^2<\infty$ tende a diventare normale. Ovvero $P_N(S_N)\rightarrow\frac{1}{\sqrt(2\pi\sigma^2_N}exp\bigg[-\frac{(S_N-\mu N)^2}{2\sigma^2_N}\bigg]$ con $\sigma^2_N=N\sigma^2$ e $\langle S_N\rangle=\mu N$, rispettando così il Teorema del Limite Centrale.

Se invece campioniamo la $S_N$ di random variabili distribuite seconda Cauchy-Lorentz si ottiene:

```python
def lorentzian( x, x0, a, gam ):
    return a * gam**2 / ( gam**2 + ( x - x0 )**2)

x1,y1= np.loadtxt("Es1.1/Histo_1.txt", usecols=(2,3), delimiter='\t', unpack='true')
x2,y2= np.loadtxt("Es1.1/Histo_2.txt", usecols=(2,3), delimiter='\t', unpack='true')
x3,y3= np.loadtxt("Es1.1/Histo_10.txt", usecols=(2,3), delimiter='\t', unpack='true')
x4,y4= np.loadtxt("Es1.1/Histo_100.txt", usecols=(2,3), delimiter='\t', unpack='true')

par,pcov = optimize.curve_fit(gaus,x4,y4,p0=[1,1.,0.2])
parl,pcovl=optimize.curve_fit(lorentzian,x4,y4,p0=[1,0.,1])

f, axarr = plt.subplots(2,2,figsize=(20,20))
f.subplots_adjust(hspace=0.2, wspace=0.2)

axarr[0, 0].bar(x1, y1, width=0.2)
axarr[0, 0].set_title(r'$N=1$',fontsize='x-large')
axarr[0, 1].bar(x2, y2, width=0.2)
axarr[0, 1].set_title(r'$N=2$',fontsize='x-large')
axarr[1, 0].bar(x3, y3, width=0.2)
axarr[1, 0].set_title(r'$N=10$',fontsize='x-large')
axarr[1, 1].bar(x4, y4, width=0.2,label="Data")
axarr[1, 1].plot(x4,gaus(x4,par[0],par[1],par[2]),'r',label="Fit Gaus")
axarr[1, 1].plot(x4,lorentzian(x4,parl[0],parl[1],parl[2]),'k',label="Fit Lorentzian")
axarr[1, 1].legend(fontsize='x-large')
axarr[1, 1].set_title(r'$N=100$',fontsize='x-large')

f.suptitle('Lorentzian Dice',fontsize='x-large')


print("Fit Gaussiana= ", par)
print("Fit Lorentziana= ", parl)
plt.show()
```

A differenza delle distribuzioni precendeti, per random variabili distribuite secondo Cauchy-Lorentz, la distribuzione limite di $S_N$ non approccia una gaussiana ma rimane una distribuzione di Cauchy-Lorentz. Infatti tale distribuzione non soddisfa le condizion del Teorema del Limite Centrale avendo varianza nfinita ma risulta essere una distribuzione stabile.

The probability $P$ that the needle will intersect one of these lines is: $P = 2L/\pi d$. This could be used to evaluate $\pi$ from throws of the needle: if the needle is thrown down $N_{thr}$ times and is observed to land on a line $N_{hit}$ of those times, we can make an estimate of $\pi$ from
